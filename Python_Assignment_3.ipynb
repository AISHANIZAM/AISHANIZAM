```python
def factorial(n):
    factorial_value = 1
    for i in range(1, n + 1):
        factorial_value *= i
    return factorial_value

# Example usage:
number = 5
print(f"The factorial of {number} is:", factorial(number))
```

```python
def calculate_simple_interest(principal, rate, time):
    # Calculate simple interest
    simple_interest = (principal * rate * time) / 100
    return simple_interest

# Example usage:
principal_amount = 1000   # Principal amount
interest_rate = 3.5      # Annual interest rate
time_period = 2          # Time period in years

# Calculate simple interest
interest = calculate_simple_interest(principal_amount, interest_rate, time_period)

# Print the result
print(f"Principal Amount: ${principal_amount}")
print(f"Interest Rate: {interest_rate}% per annum")
print(f"Time Period: {time_period} years")
print(f"Simple Interest: ${interest}")
```

```python
def calculate_compound_interest(principal, rate, time, compounds_per_year):
    # Calculate compound interest
    amount = principal * (1 + rate / compounds_per_year) ** (compounds_per_year * time)
    compound_interest = amount - principal
    return compound_interest

# Example usage:
principal_amount = 1000   # Principal amount
interest_rate = 3.5       # Annual interest rate (as a decimal)
time_period = 2           # Time period in years
compounds_per_year = 12   # Compounded monthly

# Calculate compound interest
interest = calculate_compound_interest(principal_amount, interest_rate/100, time_period, compounds_per_year)

# Print the result
print(f"Principal Amount: ${principal_amount}")
print(f"Interest Rate: {interest_rate}% per annum")
print(f"Time Period: {time_period} years")
print(f"Compounds per year: {compounds_per_year}")
print(f"Compound Interest: ${interest:.2f}")
```

```python
def is_armstrong_number(number):
    """Checks if a number is an Armstrong number."""
    num_str = str(number)
    num_digits = len(num_str)
    sum_of_powers = sum(int(digit) ** num_digits for digit in num_str)
    return sum_of_powers == number

# Example usage:
number = 153
if is_armstrong_number(number):
    print(f"{number} is an Armstrong number")
else:
    print(f"{number} is not an Armstrong number")
```

```python
import math

def calculate_area_of_circle(radius):
    # Calculate the area of the circle
    area = math.pi * (radius ** 2)
    return area

# Example usage:
radius = 3.5  # Radius of the circle

# Calculate area
area = calculate_area_of_circle(radius)

# Print the result
print(f"The area of the circle with radius {radius} is: {area:.2f}")
```

```python
def is_prime(num):
    """ Function to check if a number is prime """
    if num <= 1:
        return False
    if num == 2:
        return True  # 2 is a prime number
    if num % 2 == 0:
        return False  # Even numbers greater than 2 are not prime
    for i in range(3, int(num**0.5) + 1, 2):
        if num % i == 0:
            return False
    return True

def print_primes_in_interval(start, end):
    """ Function to print all prime numbers within an interval [start, end] """
    print(f"Prime numbers between {start} and {end}:")
    for num in range(start, end + 1):
        if is_prime(num):
            print(num, end=" ")

# Example usage:
start_num = 10
end_num = 50
print_primes_in_interval(start_num, end_num)
```

```python
def fibonacci(n):
    # Base cases
    if n <= 0:
        return 0
    elif n == 1:
        return 1

    # Initialize variables to store the two previous Fibonacci numbers
    a, b = 0, 1

    # Iterate to find the n-th Fibonacci number
    for _ in range(2, n + 1):
        # Calculate the next Fibonacci number by adding the previous two
        a, b = b, a + b

    return b

# Example usage:
n = 10
print(f"The {n}-th Fibonacci number:", fibonacci(n))
```

```python
import math

def is_perfect_square(x):
    # Check if x is a perfect square
    s = int(math.sqrt(x))
    return s * s == x

def is_fibonacci(num):
    # Check if num is a Fibonacci number
    if num <= 0:
        return False
    # Check both conditions for Fibonacci numbers
    return is_perfect_square(5 * num * num + 4) or is_perfect_square(5 * num * num - 4)

# Example usage:
number = 13  # Change this number to check different cases
if is_fibonacci(number):
    print(f"{number} is a Fibonacci number")
else:
    print(f"{number} is not a Fibonacci number")
```

```python
def fibonacci_nth_multiple(number, n):
    fib1, fib2 = 0, 1  # Initialize first two Fibonacci numbers
    count = 0  # Count of multiples found

    while count < n:
        fib_next = fib1 + fib2  # Calculate the next Fibonacci number
        if fib_next % number == 0:  # Check if it's a multiple of the given number
            count += 1
            if count == n:
                return fib_next
        fib1, fib2 = fib2, fib_next  # Move to the next Fibonacci numbers

    return None  # In case no such multiple is found (though Fibonacci sequence has infinite numbers)

# Example usage:
number = 3   # The number whose multiples we are interested in
n = 5        # The position (nth) of the multiple in the Fibonacci sequence

result = fibonacci_nth_multiple(number, n)
if result:
    print(f"The {n}-th multiple of {number} in Fibonacci series is:", result)
else:
    print(f"No such multiple found in the first {n} Fibonacci numbers.")
```

```python
def sum_of_squares(n):
    sum = 0
    for i in range(1, n + 1):
        sum += i ** 2
    return sum

# Example usage:
n = 5
result = sum_of_squares(n)
print(f"The sum of squares of first {n} natural numbers is:", result)
```

```python
def cube_sum(n):
    sum = 0
    for i in range(1, n + 1):
        sum += i ** 3
    return sum

# Example usage:
n = 5
result = cube_sum(n)
print(f"The sum of cubes of first {n} natural numbers is:", result)
```
